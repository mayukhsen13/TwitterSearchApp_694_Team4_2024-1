## Tweet containing a string

```python
def string_search(words, choice_and_or, date_and_or, range_choice, time_and_or, time_choice):
    
    if choice_and_or == "AND":
        text_where_clause = "AND ".join([f"text LIKE '% {word} %'" for word in words])
    elif choice_and_or == "OR":
        text_where_clause = "OR ".join([f"text LIKE '% {word} %'" for word in words])
    else: 
        text_where_clause = (f"text LIKE '% {words} %'")
        
    if date_and_or == "AND RANGE" or date_and_or == "SINGLE RANGE":
        date_where_clause = "AND ".join([f"DATE(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) {condition}" for condition in range_choice])
    elif date_and_or == "OR RANGE":
        date_where_clause = "OR ".join([f"DATE(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) {condition}" for condition in range_choice])
    else: 
        date_where_clause = f"DATE(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) {range_choice}"

    if time_and_or == "AND RANGE" or time_and_or == "SINGLE RANGE":
        time_where_clause = "AND ".join([f"TIME(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) {condition}" for condition in time_choice])
    elif time_and_or == "OR RANGE":
        time_where_clause = "OR ".join([f"TIME(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) {condition}" for condition in time_choice])
    else: 
        time_where_clause = f"TIME(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) {time_choice}"


    where_clauses = [text_where_clause]

    if range_choice:
        where_clauses.append(date_where_clause)

    if time_choice:
        where_clauses.append(time_where_clause)

    if len(where_clauses) > 1:
        where_clause = " AND ".join(where_clauses)
    else:
        where_clause = text_where_clause


    sql_query = f"""
    SELECT 
        text, 
        retweet_count, 
        DATE(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) AS date,
        TIME(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) AS time,
        id_str_user, 
        id_str_tweet
    FROM `msds-417117.Tweets.Tweets`
    WHERE {where_clause}
    ORDER BY retweet_count DESC
    """

    query = client.query(sql_query)
    result = query.to_dataframe()

    return result
```

```python
import heapq
import time
import pickle
from IPython.display import clear_output, display
```

```python
class SearchCache:
    def __init__(self, max_size=35):
        self.max_size = max_size
        self.heap = []
        self.cache = {}

    def store_search(self, search_input, search_result):
        # If cache is full, remove the oldest search
        if len(self.cache) >= self.max_size:
            oldest_search = heapq.heappop(self.heap)[1]
            del self.cache[oldest_search]

        # Add the new search input and result
        self.cache[search_input] = search_result
        heapq.heappush(self.heap, (time.time(), search_input))

    def get_search_result(self, search_input):
        # If search input is found in cache, return the result
        if search_input in self.cache:
            return self.cache[search_input]
        else:
            return None

    def save_cache(self, filename):
        with open(filename, "wb") as file:
            pickle.dump(self.cache, file)

    def load_cache(self, filename):
        try:
            with open(filename, "rb") as file:
                self.cache = pickle.load(file)
                # Reconstruct the heap from the loaded cache
                self.heap = [(time.time(), key) for key in self.cache]
                heapq.heapify(self.heap)
        except FileNotFoundError:
            print("Cache file not found. Starting with an empty cache.")

# Create an instance of the cache
search_cache = SearchCache()
search_cache.load_cache("/Users/divyashah/Downloads/Twitter project/search_cache.pkl")

def search():
    clear_output(wait=True)
    
    choice = input('''Which metric would you like to search? Input the number corresponding to your choice.
    1 -> Search by word in tweet
    2 -> Search by hashtag in tweet
    3 -> Search by number of retweets
    4 -> RESTART
    5 -> EXIT
    ''' + "\n")
    
    if choice == "5":
        print("Exiting the program.")
        return
    
    if choice == "4" or choice not in ("1", "2", "3"):
        search()
    
    date_and_or = input("\n" + '''Would you like to provide a date range to search? Type 'AND RANGE', 'OR RANGE', 'SINGLE RANGE', or 'NO'.
    ''' + "\n")
    
    if date_and_or == "AND RANGE" or date_and_or == "OR RANGE":
        range_input = input("\n" + '''Enter your conditions on date in exact format without brackets as [ {operator} 'YYYY-MM-DD' ]. 
        Separated by commas, use exactly the symbols: = , < , > , >= , <=
        Example: [ < '2020-04-25', > '2020-03-07' ]\n''' + "\n")
        range_choice = [condition.strip() for condition in range_input.split(',')]
    elif date_and_or == "SINGLE RANGE":
        range_input = input("\n" + '''Enter your condition on date in exact format without brackets as [ {operator} 'YYYY-MM-DD' ]. 
        Use exactly the symbols: = , < , > , >= , <=
        Example: [ < '2020-04-25' ]\n''' + "\n")
        range_choice = [range_input]
    else:
        range_choice = ''
        
    time_and_or = input("\n" + '''Would you like to provide a time range to search? Type 'AND RANGE', 'OR RANGE', 'SINGLE RANGE', or 'NO'.
    ''' + "\n")
        
    if time_and_or == "AND RANGE" or time_and_or == "OR RANGE":
        time_input = input("\n" + '''Enter your conditions on time in exact format without brackets as [ {operator} '24HR:MIN:SEC' ]. 
        Separated by commas, use exactly the symbols: = , < , > , >= , <=
        Example: [ < '13:00:00', > '12:40:00' ]\n''' + "\n")
        time_choice = [condition.strip() for condition in time_input.split(',')]
    elif time_and_or == "SINGLE RANGE":
        time_input = input("\n" + '''Enter your condition on time in exact format without brackets as [ {operator} '24HR:MIN:SEC' ]. 
        Use exactly the symbols: = , < , > , >= , <=
        Example: [ < '13:00:00' ]\n''' + "\n")
        time_choice = [time_input]
    else:
        time_choice = ''
        
    if choice == "1":
        choice_and_or = input("\n" + '''Type 'AND' if you want to search for tweets containing all words. 
Type 'OR' if you want to search for tweets containing any of the words.
Type any key to search for one word or phrase.\n''' + "\n")
        
        if choice_and_or == "AND" or choice_and_or == "OR":
            words_input = input("\nEnter the words for which you'd like to search, separated by commas.\n")
            words = [word.strip() for word in words_input.split(',')]
        else:
            words_input = input("\nEnter the word or phrase for which you'd like to search.\n")
            words = words_input
        
        # Check cache for search input
        cache_key = (words, choice_and_or, date_and_or, range_choice, time_and_or, time_choice)
        cached_result = search_cache.get_search_result(cache_key)
        
        if cached_result is not None:
            print("\n" + "################################################")
            print("\n" + "Returning result from cache:")
            display(cached_result)
            explore_request(cached_result)
            return
        
        result = string_search(words, choice_and_or, date_and_or, range_choice, time_and_or, time_choice)
        
        if result.empty:
            print("\n" + "################################################")
            print("\n" + "No matching tweets. Please continue your search.")
            search()
        else:
            search_cache.store_search(cache_key, result)
            display(result)
            explore_request(result)
        

    if choice == "2":
        choice_and_or = input("\n" + '''Type 'AND' if you want to search for tweets containing all hashtags. 
Type 'OR' if you want to search for hashtags containing any of the hashtags.
Type any key to search for one hashtag.\n''' + "\n")
        
        if choice_and_or == "AND" or choice_and_or == "OR":
            hashtags_input = input("\nEnter the hashtags for which you'd like to search, separated by commas.\n")
            hashtags = [hashtag.strip() for hashtag in hashtags_input.split(',')]
        else:
            hashtags_input = input("\nEnter the hashtag for which you'd like to search.\n")
            hashtags = hashtags_input
        
        # Check cache for search input
        cache_key = (hashtags, choice_and_or, date_and_or, range_choice, time_and_or, time_choice)
        cached_result = search_cache.get_search_result(cache_key)
        
        if cached_result is not None:
            print("\n" + "################################################")
            print("\n" + "Returning result from cache:")
            display(cached_result)
            explore_request(cached_result)
            return
        
        result = hashtag_search(hashtags, choice_and_or, date_and_or, range_choice, time_and_or, time_choice)
        
        if result.empty:
            print("\n" + "################################################")
            print("\n" + "No matching tweets. Please continue your search.")
            search()
        else:
            search_cache.store_search(cache_key, result)
            display(result)
            explore_request(result)


    if choice == "3":
        choice_and_or = input("\n" + '''Type 'AND' if you want to search for tweets by all conditions on retweets. 
Type 'OR' if you want to search for tweets by any condition on retweets.
Type any key to search for tweets with a single condition on retweets.\n''' + "\n")
        
        if choice_and_or == "AND" or choice_and_or == "OR":
            retweets_input = input("\n" + '''\nEnter the conditions on retweets for which you'd like to search.
        Separated by commas, use exactly the symbols: = , < , > , >= , <=
        Example: [ > 215, < 300 ]\n''' + "\n")
            retweets = [conditions.strip() for conditions in retweets_input.split(',')]
        else:
            retweets_input = input("\n" + '''\nEnter the condition on retweets for which you'd like to search.
        Use exactly the symbols: = , < , > , >= , <=
        Example: [ > 215 ]\n''' + "\n")
            retweets = retweets_input
        
        # Check cache for search input
        cache_key = (retweets, choice_and_or, date_and_or, range_choice, time_and_or, time_choice)
        cached_result = search_cache.get_search_result(cache_key)
        
        if cached_result is not None:
            print("\n" + "################################################")
            print("\n" + "Returning result from cache:")
            display(cached_result)
            explore_request(cached_result)
            return
        
        result = retweet_search(retweets, choice_and_or, date_and_or, range_choice, time_and_or, time_choice)
        
        if result.empty:
            print("\n" + "################################################")
            print("\n" + "No matching tweets. Please continue your search.")
            search()
        else:
            search_cache.store_search(cache_key, result)
            display(result)
            explore_request(result)

    # Save the cache after updating it
    search_cache.save_cache("/Users/divyashah/Downloads/Twitter project/search_cache.pkl")
    
    explore_request(result)
            
def explore_request(result):
    
    explore_confirm = input('''Would you like to explore tweets further? Input the number corresponding to your search.
    1 -> Find all retweeting users
    2 -> Find all author's tweets
    3 -> Show top 10 tweets by retweet
    4 -> RESTART
    5 -> EXIT
    ''')
    
    if explore_confirm == "5":
        print("Exiting the program.")
        return
    
    if explore_confirm == "4":
        search()
    
    if explore_confirm == "1":
        clear_output(wait=True)
        explore_tweet_id = input("\n" + '''Enter the id_str_tweet of the tweet for which you'd like to find all retweeting users.''' + "\n")
        explore_retweeting_users(explore_tweet_id)
        
    if explore_confirm == "2":
        clear_output(wait=True)
        explore_user_id = input("\n" + '''Enter the id_str_user of the user for which you'd like to find all tweets.''' + "\n")
        explore_authors_tweets(explore_user_id)
    
    if explore_confirm == "3":
        clear_output(wait=True)
        display(result.head(10))
        explore_request(result.head(10))

def explore_retweeting_users(tweet_id):
    
    sql_query = f"""
    SELECT 
        text, 
        retweet_count, 
        DATE(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) AS date,
        TIME(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) AS time,
        id_str_retweet_user,
        id_str_retweet
    FROM `msds-417117.Retweets.Retweets`
    WHERE id_str_tweet = '{tweet_id}'
    ORDER BY retweet_count DESC
    """

    query = client.query(sql_query)
    result = query.to_dataframe()

    if result.empty:
        clear_output(wait=True)
        print("\n" + "################################################")
        print("\n" + "No matching tweets. Please continue your search.")
        search()
    else:
        display(result)
        explore_request(result)
        return result


def explore_authors_tweets(user_id):
    
    sql_query = f"""
    SELECT 
        text, 
        retweet_count, 
        DATE(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) AS date,
        TIME(PARSE_TIMESTAMP('%a %b %d %T %z %Y', created_at)) AS time,
        id_str_user,
        id_str_tweet
    FROM `msds-417117.Tweets.Tweets`
    WHERE id_str_user = '{user_id}'
    ORDER BY retweet_count DESC
    """

    query = client.query(sql_query)
    result = query.to_dataframe()
    
    if result.empty:
        clear_output(wait=True)
        print("\n" + "################################################")
        print("\n" + "No matching tweets. Please continue your search.")
        search()
    else:
        display(result)
        explore_request(result)
        return result

# Call the search function to start searching
search()
search_cache.save_cache("/Users/divyashah/Downloads/Twitter project/search_cache.pkl")
```
